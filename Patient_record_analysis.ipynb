```python
numeric_cols = ['Age', 'Cost', 'Length_of_Stay' , 'Satisfaction']

# Set style for plots
sns.set(style="whitegrid")

# Loop through numeric columns and create boxplots
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=p_record[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

## Univariate Analysis of the data

```python
numeric_cols = ['Age','Length_of_Stay']

# Set style for plots
sns.set(style="whitegrid")

# Loop through numeric columns
for col in numeric_cols:
    # ----- Histogram -----
    plt.figure(figsize=(8, 4))
    sns.histplot(p_record[col], kde=True, bins=10, color='skyblue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
hist_cost = [ 'Cost']

# Set style for plots
sns.set(style="whitegrid")

# Loop through numeric columns
for col in hist_cost:
    # ----- Histogram -----
    plt.figure(figsize=(8, 4))
    sns.histplot(p_record[col], kde=True, bins=8, color='skyblue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
hist_satisfacton = ['Satisfaction']

# Set style for plots
sns.set(style="whitegrid")

# Loop through numeric columns
for col in hist_satisfacton :
    # ----- Histogram -----
    plt.figure(figsize=(8, 4))
    sns.histplot(p_record[col], kde=True, bins=4, color='skyblue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
numeric_cols = ['Age', 'Cost', 'Length_of_Stay', 'Satisfaction']

sns.set(style="whitegrid")

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.kdeplot(p_record[col], shade=True, color='orange')
    plt.title(f'Density Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

## Univariate analsis of the catagorical variable

```python
# Categorical columns
categorical_cols = ['Gender', 'Disease', 'Treatment', 'Readmitt', 'Outcome']

# Set style
sns.set(style="whitegrid")

for col in categorical_cols:
    # ----- Countplot -----
    plt.figure(figsize=(8, 4))
    sns.countplot(x=p_record[col], palette='pastel', order=p_record[col].value_counts().index)
    plt.title(f'Countplot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=90)  # rotate labels if long
    plt.show()
```

```python
# Categorical columns
categorical_cols = ['Disease', 'Treatment', 'Readmitt', 'Outcome']

# Set style
sns.set(style="whitegrid")

for col in categorical_cols:
# ----- Pie Chart -----
    plt.figure(figsize=(6, 6))
    p_record[col].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')  # remove y-label
    plt.show()
```

```python
avg_cost = p_record.groupby('Disease')['Cost'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=avg_cost, x='Disease', y='Cost')
plt.xticks(rotation=45)
plt.title('Average Cost per Disease')
plt.ylabel('Average Cost')
plt.xlabel('Disease')
plt.show()
```

```python
# Plot ki size set karein
plt.figure(figsize=(15, 6))

# Countplot use karein: X-axis par Age aur 'hue' mein Disease
sns.countplot(data=p_record, x='Age', hue='Disease')

plt.title('Age vs Disease Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.legend(title='Disease Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45) # Agar ages zyada hain toh rotate kar dein
plt.show()
```

```python
# Plot  size
plt.figure(figsize=(12, 6))

# Countplot: x shows Disease  and hue (color)  Gender
sns.countplot(data=p_record, x='Disease', hue='Gender', palette='Set2')

plt.title('Gender Distribution across Different Diseases')
plt.xlabel('Disease Name')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45) # Disease name
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(data=p_record, x='Disease', hue='Outcome', palette='magma')
plt.title('Analysis of Disease vs Patient Outcome', fontsize=15)
plt.xlabel('Disease Type', fontsize=12)
plt.ylabel('Patient Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Outcome Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Set the visual style of the plots
sns.set_theme(style="whitegrid")

# Create a figure to visualize Age distribution by Outcome
plt.figure(figsize=(12, 6))

# Using a Histogram with 'kde=True' to show the density and count of ages
sns.histplot(data=p_record, x='Age', hue='Outcome', kde=True, element="step", palette='viridis')

# Adding professional titles and axis labels
plt.title('Age Distribution by Patient Outcome', fontsize=15)
plt.xlabel('Patient Age', fontsize=12)
plt.ylabel('Frequency / Patient Count', fontsize=12)

# Adjusting the layout for a clean look
plt.tight_layout()

# Display the plot
plt.show()
```

```python


# Set the figure size for clarity
plt.figure(figsize=(12, 6))

# Use a Regression Plot to see the trend between Age and Satisfaction
# This shows individual data points and a trend line
sns.regplot(data=p_record, x='Age', y='Satisfaction', 
            scatter_kws={'alpha':0.3}, # Make dots slightly transparent
            line_kws={'color':'red'})  # Trend line in red for visibility

# Adding professional titles and labels
plt.title('Correlation between Patient Age and Satisfaction Score', fontsize=15)
plt.xlabel('Patient Age', fontsize=12)
plt.ylabel('Satisfaction Level (1-5)', fontsize=12)

# Improve the overall look
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot size for better visibility of 3-feature relationship
plt.figure(figsize=(16, 8))

# We use 'Disease' on X-axis, 'Treatment Cost' on Y-axis
# 'hue' is used to split the bars by 'Treatment' type
sns.barplot(data=p_record, x='Disease', y='Cost', hue='Treatment', palette='tab10')

# Adding professional English titles and labels
plt.title('Analysis of Treatment Costs by Disease and Procedure Type', fontsize=16)
plt.xlabel('Disease Category', fontsize=12)
plt.ylabel('Treatment Cost ($)', fontsize=12)

# Rotating X-axis labels to prevent overlap
plt.xticks(rotation=45, ha='right')

# Placing the legend outside because there are many treatment types
plt.legend(title='Specific Treatment', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Set the visual style
sns.set_theme(style="whitegrid")

# Create a figure to compare Age distribution for Readmitted vs Non-Readmitted patients
plt.figure(figsize=(10, 6))

# Using a Boxplot to show the median age and spread for each Readmission status
sns.boxplot(data=p_record, x='Readmitt', y='Age', palette='Set2')

# Adding professional English titles and labels
plt.title('Age Distribution by Readmission Status', fontsize=15)
plt.xlabel('Readmitted (Yes/No)', fontsize=12)
plt.ylabel('Patient Age', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
treatment_total_cost = p_record.groupby('Treatment')['Cost'].sum().sort_values(ascending=False).reset_index()

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a bar plot showing the total expenditure per treatment
# Sorting the bars makes it easier to identify the most expensive treatments
sns.barplot(data=treatment_total_cost, x='Cost', y='Treatment', palette='viridis')

# Adding professional titles and labels
plt.title('Total Expenditure by Treatment Type', fontsize=16)
plt.xlabel('Total Cost (Sum)', fontsize=12)
plt.ylabel('Treatment Description', fontsize=12)

# Adding grid lines for better readability of the cost scale
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style for a professional portfolio look
sns.set_theme(style="whitegrid")

# Create a figure to analyze Satisfaction across Gender and Disease
plt.figure(figsize=(15, 8))

# Using a Point Plot to show the average satisfaction
# This clearly shows the difference in satisfaction between Male and Female for each Disease
sns.pointplot(data=p_record, x='Disease', y='Satisfaction', hue='Gender', 
              markers=["o", "s"], linestyles=["-", "--"], palette='Set1')

# Adding professional English titles and labels
plt.title('Average Patient Satisfaction by Gender and Condition', fontsize=16)
plt.xlabel('Medical Condition (Disease)', fontsize=12)
plt.ylabel('Average Satisfaction Score (1-5)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Position the legend
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
